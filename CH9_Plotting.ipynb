```python
# import pyplot from matplotlib with an alias
import matplotlib.pyplot as plt
```

```python
# first, we'll take the simplest case and plot two lists of equal length

x = [1, 2, 3, 4, 5]
y = [2, 4, 6, 8, 10]

plt.plot(x, y)
plt.xlabel('Numbers')
plt.ylabel('Doubles')
```

```python
# next, we'll plot directly from a pandas dataframe
import pandas

data = pandas.read_csv('data/gapminder_gdp_oceania.csv', index_col='country')
data.loc['Australia'].plot()

# note that without the 90 rotation, our x-ticks will run into (over write) each other on the x-axis 
# to see this happen, remove or comment out the next line of code
plt.xticks(rotation=90)
```

```python
# next, let's plot all the data in the dataframe
# note - we need to take the Transpose of this dataframe to
# plot the GDP label on the x-axis and the 
# value on the y-axis

data.T.plot()
plt.ylabel('GDP per capita')
plt.xticks(rotation=90)
```

```python
# different styles are available.
# for more info, see the pyplot documentation
# here's an example for using bar charts

plt.style.use('ggplot')
data.T.plot(kind='bar')
plt.xticks(rotation=90)
plt.ylabel('GDP per capita')
```

```python
# you might want to shorten the x-labels to only list the year
# ie./, 1952 instead of gdpPercap_1952

# Accumulator pattern to collect years (as character strings).

# note that we don't need to rotate this 90 degrees anymore
# though this is still an option if you prefer. 

years = []
for col in data.columns:
    year = col[-4:]
    years.append(year)

# Select two countries' worth of data.
gdp_australia = data.loc['Australia']
gdp_nz = data.loc['New Zealand']

# Plot with differently-colored markers.
plt.plot(years, gdp_australia, 'b-', label='Australia')
plt.plot(years, gdp_nz, 'g-', label='New Zealand')

# Create legend.
plt.legend(loc='upper left')
plt.xlabel('Year')
plt.ylabel('GDP per capita ($)')
```

```python
# instead of a time series, you might be interested in the correlation between two variables
# to do this, you can plot one series on the x-axis, and one on the y-axis
plt.scatter(gdp_australia, gdp_nz)
```
