```python
# ADDENDUM: Pandas Sorting, Grouping, and Aggregation
#
# Objectives
#   Sort a Pandas Dataframe by one or more columns in ascending or descending order
#   Calculate count, min, max, average, or other aggregate operations by splitting data into groups
#   based on a common value or set of values
```

```python
import pandas
```

```python
data = pandas.read_csv('data/gapminder_all.csv')
```

```python
display(data)
```

```python
# Sorting
#
# Sort a dataframe by providing a list of columns
# By default, pandas will sort in ascending order
```

```python
data.sort_values(['country'])
```

```python
# Multiple Sort
#
# Sort by multiple columns. Sorting will be applied in the order listed
# You can specify an optional ordering list - 0 will sort in ascending order, 1 will sort in descending order.
#
# The example below will sort first by continent, then by country name, both in ascending order
```

```python
data.sort_values(['continent', 'country'], ascending=[0, 0])
```

```python
# By default, pandas will display only a limited number of rows and columns to save space.
# You can set these values through the pandas.set_option method
# To display everything, set a max of "None" on max_columns and max_rows
# Note that can run out of memory if you do this with a very large dataset!
```

```python
pandas.set_option('display.max_columns', None)
pandas.set_option('display.max_rows', None)
```

```python
# to avoid displaying too much data, I'll limit this to the first 10
display(data).head(10)
```

```python
# Grouping
#
# We can also create summary statistics by group
# For example, we might want to know the mean, standard deviation, and other data
# organized by Continent.
# Note that you can group by multiple columns - do do this, pass a list of column namnes
```

```python
by_continent = data.groupby(['continent'])
```

```python
# note that once you have grouped your data, you can no longer see it on a row-by-row basis.
# you'll need to describe the data frame to see results
by_continent.describe()
```

```python
# limit results to only one column
by_continent['gdpPercap_2007'].describe()
```

```python
# Other grouping operations are available
# The example below takes the sum of each column grouped by continent
```

```python
by_continent.sum()
```
