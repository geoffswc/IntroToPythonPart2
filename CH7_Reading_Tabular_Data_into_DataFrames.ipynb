```python
# use index_col to specify that a column's values should be used as row headings

# In the table above, the first column contains the row headings but we would like the row heading to be country 

# We pass the name of the column to read_csv method as its index_col parameter
```

```python
data = pd.read_csv('data/gapminder_gdp_oceania.csv', index_col='country')
```
